### 1.1. What's a GeoDataFrame?

We used the GeoPandas library to read in the geospatial data, and this returned us a `GeoDataFrame`:

```python
# we can check the type of the 'countries' data by running type(countries)
type(countries)
```

A GeoDataFrame contains a tabular, geospatial dataset:

* It has a **'geometry' column** that holds the geometry information.
* The other columns are the **attributes** that describe each of the geometries.

Such a `GeoDataFrame` is just like a pandas `DataFrame`, but with some additional functionality for working with geospatial data:

* A `.geometry` attribute that always returns the column with the geometry information (returning what is called a 'GeoSeries'). The column name itself does not necessarily need to be 'geometry', but it will always be accessible as the `.geometry` attribute.
* It has some extra methods for working with spatial data (area, distance, buffer, intersection, ...), which we will see in later notebooks

```python
# We can isolate the geometry column for a closer look
countries.geometry
```

```python
# A second notation type is also available:
countries['geometry']
```

We can use the operations that Pandas provides to manipulate the geospatial dataset. For instance, the .area operation allows for the calculation of the area of each polygon. Similarly, we can use .mean() to calculate the average of an attribute in the data. We can also filter the data to get, for example, only the countr(y)(ies) of interest.

### 1.3. Coordinate reference systems

As briefly mentioned above, a **coordinate reference system (CRS)** determines how the two-dimensional (planar) coordinates of the geometry objects should be related to actual places on the (non-planar) earth.

For a nice in-depth explanation, see https://docs.qgis.org/2.8/en/docs/gentle_gis_introduction/coordinate_reference_systems.html

A GeoDataFrame or GeoSeries has a `.crs` attribute which holds (optionally) a description of the coordinate reference system of the geometries:

```python
# check the coordinate reference system of the countries dataset
countries.crs
```

For the `countries` dataframe, it indicates that it used the EPSG 4326 / WGS84 lon/lat reference system, which is a common CRS.  
It uses coordinates as latitude and longitude in degrees, as can you be seen from the x/y labels on the plot we made in the beginning.

The `.crs` attribute is given as a dictionary. Under the hood, GeoPandas uses the `pyproj` / `proj4` libraries to deal with the re-projections.

For more information, see also http://geopandas.readthedocs.io/en/latest/projections.html.

The epsg 4326 crs can be referred to via http://epsg.io/. For example, if you navigate to http://epsg.io/4326, you will find the description of this projection:



---

There are sometimes good reasons you want to change the coordinate references system of your dataset, for example:

- different sources with different crs -> need to convert to the same crs
- distance-based operations -> you will need a crs that is in, for example, meters (not degrees)
- plotting in a certain crs (eg to preserve area)

We can convert a GeoDataFrame to another reference system using the `to_crs` function. 

For example, let's convert the countries to the World Mercator projection (http://epsg.io/3395):

```python
# remove Antartica, as the Mercator projection cannot deal with the poles

# the following code takes all entries with NAME that does not equal Antarctica and calls it 'countries'
# the old 'countries' dataset is now no longer available unless reloaded as we did in the beginning
countries = countries[(countries['NAME'] != "Antarctica")]
```

```python
# convert shapefile crs to mercator projection and call it 'countries_mercator'
countries_mercator = countries.to_crs(epsg=3395)  # or .to_crs({'init': 'epsg:3395'})
```

```python
# Plot the countries_mercator data
countries_mercator.plot()
```

As you can see, the data is now illustrated in a different projection compared to the first plot of the world. Relationships about space and distance are now represented differently. 

```python
# calculate area of each country in countries_mercator depiction
countries_mercator.area
```

**Note that the area numbers are entirely different than reported earlier.**

### 1.5. Relationships between Spatial Data

An important aspect of geospatial data is that we can look at *spatial relationships*: how two spatial objects relate to each other (whether they overlap, intersect, contain, .. one another).

The topological, set-theoretic relationships in GIS are typically based on the DE-9IM model. See https://en.wikipedia.org/wiki/Spatial_relation for more information.

![](img/TopologicSpatialRelations2.png)


(Image by [Krauss, CC BY-SA 3.0](https://en.wikipedia.org/wiki/Spatial_relation#/media/File:TopologicSpatialRelarions2.png))

To show how these relationships can be tested using Geopandas, let's determine the distance between Seoul, South Korea and Pyongyang, North Korea. 

```python
# get the polygon area of South Korea and Chine
South_Korea = countries.loc[countries['NAME'] == 'South Korea', 'geometry'].squeeze()
China = countries.loc[countries['NAME'] == 'China', 'geometry'].squeeze()
# note: we use `.squeeze()` here to to extract the scalar geometry object from the GeoSeries of length 1)
```

Next, let's reload Seoul and Beijing.

```python
Seoul = cities.loc[cities['name'] == 'Seoul', 'geometry'].squeeze()
Beijing = cities.loc[cities['name'] == 'Beijing', 'geometry'].squeeze()
```

Now, we show a line between Seoul and Beijing.

```python
from shapely.geometry import LineString
line = LineString([Seoul, Beijing])
```

Let's visualize those 4 geometry objects together (I only put them in a GeoSeries to easily display them together with the geopandas `.plot()` method):

```python
gpd.GeoSeries([South_Korea, China, Seoul, Beijing, line]).plot(cmap='tab10')
```

You can recognize the abstract shape of South Korea and China. The line connects Seoul and Beijing. Seoul, the capital of South Korea, is thus located within South Korea. This is a spatial relationship, and we can test this using the individual shapely geometry objects as follow:

```python
# Test whether Soul is inside South Korea
Seoul.within(South_Korea)
```

```python
# Test whether South Korea contains Seoul
South_Korea.contains(Seoul)
```

```python
# Test whether South Korea contains Beijing
South_Korea.contains(Beijing)
```

```python
# Test whether South Korea contains the linestring
South_Korea.contains(line)
```

```python
# test whether linestring and South Korea intersect
line.intersects(South_Korea)
```

The same methods that are available on individual geometries as we have seen above, are also available as methods on `GeoSeries` / `GeoDataFrame` objects. For example, if we call the `contains` method on the world dataset with the `Seoul` point, it will do this spatial check for each country in the `world` dataframe:

```python
# check each country whether it contains Seoul
countries.contains(Seoul)
```

```python
# filter the countries dataset for only those countries that contain Seoul
countries[countries.contains(Seoul)]
```

### 1.6. Spatial operations

Shapely and GeoPandas aslo provide analysis methods that return new geometric objects.

See https://shapely.readthedocs.io/en/stable/manual.html#spatial-analysis-methods for more details.

For example, let's construct a buffer around Seoul (which returns a Polygon):

```python
# set buffer around Seoul and plot Seoul and South Korea
gpd.GeoSeries([South_Korea, Seoul.buffer(1)]).plot(alpha=0.5, cmap='tab10')
```

and now take the intersection, union or difference of those two polygons:

```python
# Take the intersection between the buffered area around Seoul and South Korea
Seoul.buffer(1).intersection(South_Korea)
```

```python
# combine the buffer and the polygon area of South Korea
Seoul.buffer(1).union(South_Korea)
```

```python
# Keep only the difference
Seoul.buffer(1).difference(South_Korea)
```

Another useful method is the `unary_union` attribute, which converts the set of geometry objects in a GeoDataFrame into a single geometry object by taking the union of all those geometries. For example, we can construct a single object for the African continent:

```python
# keep only those countries that are part of the African continent
africa_countries = countries[countries['CONTINENT'] == 'Africa']
```

```python
# Plot countries of Africa
africa_countries.plot()
```

```python
# Use .unary_union function to dissolve the countries of Africa into one polygon
africa = africa_countries.unary_union
```

```python
# Plot the polygon of the Africa continent
africa
```

### 1.7. Combining Data based on their Spatial Relationship

Earlier, we combined data into one visualization. However, we can also merge different dataframes based on a relationship between the datasets. Here, we illustrate first how to conduct a join based on a common attribute and, next, based on a spatial relationship.

To illustrate the concept of joining the information of two dataframes with pandas, let's take a small subset of our `cities` and `countries` datasets: 

```python
# take small subset of cities data, keeping only Bern, Brussels, London, and Paris
cities2 = cities[cities['name'].isin(['Bern', 'Brussels', 'London', 'Paris'])].copy()

# Add a column to the data for their ISO_A3 country code
cities2['ISO_A3'] = ['CHE', 'BEL', 'GBR', 'FRA']
```

```python
# Create a new countries2 dataset that only contains the three columns of the countries data that describe the country ISO_A3 code,
# the name, and the continent
countries2 = countries[['ISO_A3', 'NAME', 'CONTINENT']]
```

We added a 'ISO_A3' column to the `cities2` dataset, indicating a code of the country of the city. This country code is also present in the `countries` dataset, which allows us to merge those two dataframes based on the common column.

Joining the `cities` dataframe with `countries` will transfer extra information about the countries (the full name, the continent) to the `cities` dataframe, based on a common key:

```python
cities2.merge(countries2, on='ISO_A3')
```

**But**, for this illustrative example, we added the common column manually, it is not present in the original dataset. However, we can still know how to join those two datasets based on their spatial coordinates. We can apply such a spatial relationship automatically using a spatial join. In this case, we want to join the `cities` dataframe with the information of the `countries` dataframe, based on the spatial relationship between both datasets.

We use the [`geopandas.sjoin`](http://geopandas.readthedocs.io/en/latest/reference/geopandas.sjoin.html) function:

<div class="alert alert-info" style="font-size:120%">
<b>SPATIAL JOIN</b> = *transferring attributes from one layer to another based on their spatial relationship* <br><br>


Different parts of this operations:

<ul>
  <li>The GeoDataFrame to which we want add information</li>
  <li>The GeoDataFrame that contains the information we want to add </li>
  <li>The spatial relationship we want to use to match both datasets ('intersects', 'contains', 'within')</li>
  <li>The type of join: left or inner join</li>
</ul>

</div>

```python
# spatially join the data together and call the new dataset 'joined'
joined = gpd.sjoin(cities, countries, op='within', how='left')
```

In the spatial join operation above, we are not changing the geometries itself. We are not joining geometries, but joining attributes based on a spatial relationship between the geometries. This also means that the geometries need to at least overlap partially.

If you want to create new geometries based on joining (combining) geometries of different dataframes into one new dataframe (eg by taking the intersection of the geometries), you want an **overlay** operation.

```python
# Let's use Africa to illustrate
africa = countries[countries['CONTINENT'] == 'Africa']

# plot Africa
africa.plot()
```

```python
# establish a new cities geometry that is a buffer around each city
cities['geometry'] = cities.buffer(2)
```

```python
# Use the .overlay() function to create a new geometry
gpd.overlay(africa, cities, how='difference').plot()
```
